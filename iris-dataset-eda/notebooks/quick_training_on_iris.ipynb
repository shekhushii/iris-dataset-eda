```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_excel("iris.xlsx")
```

```python
data.head()
```

```python
x = data.iloc[: , :-1]
```

```python
x
```

```python
y = data.iloc[: , -1]
```

```python
y
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x.values,y.values)
```

```python
predictions = model.predict([[5.6,	2.9 ,	3.6	, 1.3]])
```

```python
predictions
```
